# mlxtend----
**mlxtend is a Python library that provides a collection of utility functions and classes for machine learning and data science tasks. The name "mlxtend" stands for "machine learning extensions." It is designed to work seamlessly with other popular machine learning libraries in Python, such as NumPy, pandas, and scikit-learn.**



The TransactionEncoder converts a DataFrame of transactions into a one-hot encoded format where each column represents a unique item, and each row corresponds to a transaction

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
store_data_dataset= pd.read_csv('store_data.csv')
```

```python
store_data_dataset
```

First we fill the NA values with space

```python
store_data_dataset.fillna('')
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(store_data_dataset).transform(store_data_dataset)
```

```python
# create the new dataframe
new_dataset = pd.DataFrame(te_ary,columns=te.columns_)
new_dataset
```

# Apriori

 It's used to discover interesting relationships between variables in large datasets. Specifically, Apriori is often applied to find frequent itemsets in transactional databases, where each transaction consists of a set of items.

```python
from mlxtend.frequent_patterns import apriori
frequent_items = apriori(new_dataset,min_support = 0.001, use_colnames = True)

frequent_items
```

 **Association rules mining**

---



we can take min_threshold as our choice


```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(frequent_items, metric='confidence',min_threshold = 0.07)
res
```

```python
res1 = res[['antecedents','consequents','support','confidence','lift']]
res1
```

```python
res2 = res1[res1['confidence']>=1]
res2
```
