Strings methods

```python
s = "hElloworld"
print(s.capitalize())
print(s.casefold())
print(s.lower())
print(s.upper())
print(s.count('l',0,4))
print(s.center(len(s)+4,'@'))
print(s.isalnum())  #isalnum is used to check string contain alphabets and (0-9)
print(s.index('l'))  #returns the first index   but if we don't find the character then it will raise an exception
print(s.find('l')) #returns -1 if the character is not find


#checking
print(s.islower())
print(s.isupper())
dict1  = {104:119}
print(s.translate(dict1))

s4 = str.maketrans('h','i')
print(s.translate(s4))
```

```python
# TypeCasting in boolean type
print(bool(1))
print(bool('hello'))
print(bool(""))
print(bool(False))
print(bool([]))
```
