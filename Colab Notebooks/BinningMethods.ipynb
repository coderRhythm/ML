# Binning Methods in Data Preprocessing:

Binning is a data preprocessing technique used to categorize or group data into bins or intervals. It is often employed to simplify complex datasets, discretize continuous variables, and prepare data for analysis or visualization.

There are two common binning methods:

 **1.Equal Frequency Binning**

 **2.Equal Width Binning. **



# Equal Frequency Binning:

*Equal frequency binning is a data preprocessing technique that divides a dataset into bins, where each bin contains approximately the same number of data points.*

# Equal Width Binning:

***Equal width binning is a data preprocessing technique that divides a dataset into bins, where each bin has the same width, and data points are assigned to bins based on their values.***

```python
import numpy as np

np.random.seed(42)
data_array = np.random.randint(1, 101, 20)

num_bins = 3

sorted_data = np.sort(data_array)

bin_edges_eq_freq = np.linspace(0, len(sorted_data) - 1, num_bins + 1, dtype=int)

bin_values_eq_freq = sorted_data[bin_edges_eq_freq]

binned_data_eq_freq = np.digitize(data_array, bin_values_eq_freq)

min_value = np.min(data_array)
max_value = np.max(data_array)
bin_width = (max_value - min_value) / num_bins

bin_edges_eq_width = np.arange(min_value, max_value + bin_width, bin_width)

binned_data_eq_width = np.digitize(data_array, bin_edges_eq_width)


print("Original Data:")
print(data_array)
print("\nEqual Frequency Binning:")
print(binned_data_eq_freq)
print("\nEqual Width Binning:")
print(binned_data_eq_width)
```
