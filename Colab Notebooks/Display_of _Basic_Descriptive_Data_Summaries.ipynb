```python
import pandas as pd
import matplotlib.pyplot as plt

# Common encodings to try: 'utf-8', 'ISO-8859-1', 'latin1'
file_path = 'sales_data_sample.csv'

try:
    df = pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    try:
        df = pd.read_csv(file_path, encoding='ISO-8859-1')
    except UnicodeDecodeError:
        df = pd.read_csv(file_path, encoding='latin1')
# Print the column names
print(df.columns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')

# Ensure that 'YEAR_ID' is the correct column name for years in your dataset
year_column_name = 'YEAR_ID'

# Specify the product columns to plot
product_columns = ['Classic Cars', 'Trucks and Buses']

# Plot product sales data as a multiline plot
plt.figure(figsize=(10, 6))
for product in product_columns:
    # Convert the product name to a string and use it as a label
    product_name = str(product)

    # Check if the product column exists in the DataFrame
    if product_name in df.columns:
        plt.plot(df[year_column_name], df[product_name], marker='o', label=product_name)

plt.title('Product Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')

# Ensure that 'YEAR_ID' is the correct column name for years in your dataset
year_column_name = 'YEAR_ID'

# Specify the product columns to plot
product_columns = df['PRODUCTLINE'].unique()  # Use the unique product lines as columns

# Plot product sales data as a multiline plot
plt.figure(figsize=(10, 6))
for product in product_columns:
    # Filter the DataFrame for the specific product line
    product_df = df[df['PRODUCTLINE'] == product]

    # Plot the sales data for the product line
    plt.plot(product_df[year_column_name], product_df['SALES'], marker='o', label=product)

plt.title('Product Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')

# Ensure that 'YEAR_ID' is the correct column name for years in your dataset
year_column_name = 'YEAR_ID'

# Specify the product columns to filter
selected_products = ['Trucks and Buses', 'Classic Cars']

# Filter data for the selected products
selected_data = df[df['PRODUCTLINE'].isin(selected_products)]

# Plot sales data using a bar chart
plt.figure(figsize=(10, 6))
plt.bar(selected_data['PRODUCTLINE'], selected_data['SALES'], color=['blue', 'green'])
plt.title('Sales Data for Selected Products')
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')

# Calculate total profit based on 'SALES' column
df['Total Profit'] = df['SALES']

# Plot a histogram of total profit
plt.figure(figsize=(8, 4))
plt.hist(df['Total Profit'], bins=10, color='purple', edgecolor='black')
plt.title('Histogram of Total Profit (Based on SALES)')
plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')

# Ensure that 'YEAR_ID' is the correct column name for years in your dataset
year_column_name = 'YEAR_ID'

# Filter data for the last year
last_year_data = df[df[year_column_name] == df[year_column_name].max()]

# Calculate total sale data for each product in the last year
product_sales_last_year = last_year_data.groupby('PRODUCTLINE')['SALES'].sum()

# Plot total sale data using a pie chart
plt.figure(figsize=(8, 8))
plt.pie(product_sales_last_year, labels=product_sales_last_year.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales for Each Product (Last Year)')
plt.show()
```
