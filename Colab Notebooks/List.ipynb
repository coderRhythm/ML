***Lists***

```python
l = [1,2,3,4,5]
print(l)
print(l.reverse())  #it gives the none value
print('after reverse: ')
l.reverse()
print(l)

for x in l:
  print(x)
print('add 6')
l.append(6)
print(l)
print('adding list [7,8,9]')
l1 = [7,8,9]
l.extend(l1)
print(l)

# remove the list item
print('removing 5')
l.remove(5)  #list-name.remove(item)
print(l)

# removing the last item
print('removing the last item')
l.pop()
print(l)

#removing the entire list
l1.clear()
print(l1)
```

```python
#Loop list
i=0;
while i<len(l):
  print(l[i])
  i=i+1
l1=[]
l = [1,2,3,4]
#Another method of looping a list
l1 = [x for x in l];
print(l1)
```

***Joining*** ***the*** ***list***

```python
first = [1,2,3]
second = [4,5,6]

#first method
print('using adding operation')
third = first+second
print(third)

#second method
print('using extend method')
fourth = first.copy()
fourth.extend(second)
print(fourth)

#third method
print('using for loop-append method')
fifth = first.copy()
for x in second:
  fifth.append(x)
print(fifth)
```
