# **SUBPLOT**
--plt.subplot(nrows,ncols,index)

```python
X = [1,2,3,4]
Y = [1,2,3,4]

#subplot1
plt.subplot(2,2,1)
plt.plot(X,Y,color='r')

#subplot2
plt.subplot(2,2,2)
plt.pie(X)

#subplot3
plt.subplot(2,2,3)
plt.bar(X,Y)

plt.show()
```

# **HISTOGRAM PLOT**
---we can generate random numbers using random numbers

---using np.random.randint(initial,final,totalNumbers)



```python
x_random = np.random.randint(10,60,(50))
print(x_random)
plt.hist(x_random,edgecolor='b',label='randomNumber plot')
plt.legend()
plt.show()
```
