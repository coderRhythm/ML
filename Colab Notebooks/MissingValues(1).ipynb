# **Handling Missing Values in a Dataset**
In this document, we will explore techniques for handling missing values in a dataset using Python in a Google Colab environment. We'll use the pandas library for data manipulation and the SimpleImputer class from scikit-learn for imputation.

````python
# Handling Missing Values in a Dataset

# In this guide, we will explore various techniques for handling missing values in a dataset using Python in a Google Colab environment. We'll use the `pandas` library for data manipulation and the `SimpleImputer` class from `scikit-learn` for imputation.

# ---

# **Step 1: Import Necessary Libraries**

# ```python
# Import necessary libraries
import pandas as pd
from sklearn.impute import SimpleImputer
````

````python
# Load your dataset
data = pd.read_csv('Alandi.csv')
````

# Identify Columns with Missing Values

````python
# Find columns with missing values
missing_columns = data.columns[data.isnull().any()]
````

# **Missing Values**: Data points that are not recorded or unavailable in the dataset.


````python
# Initialize an imputer with a strategy (e.g., 'most_frequent')
imputer = SimpleImputer(strategy='most_frequent')
data[missing_columns] = imputer.fit_transform(data[missing_columns])
````

**Imputation**: The process of replacing missing values with estimated or calculated values.

**SimpleImputer**: A scikit-learn class used for imputing missing values based on various strategies.

````python
# If desired, remove rows with any missing values
data.dropna(inplace=True)
````

````python
# Assuming 'numeric_column' corresponds to one of the unnamed columns (e.g., 'Unnamed: 2')
# You can access the column by its integer position
numeric_column_index = 2
numeric_column_name = data.columns[numeric_column_index]

# Check for missing values in the 'numeric_column'
missing_values_in_numeric_column = data[numeric_column_name].isnull().sum()
print(f"Missing values in '{numeric_column_name}': {missing_values_in_numeric_column}")
````

````python
# 1. Data Cleaning
# Replace any non-numeric values with NaN (missing values)
data[numeric_column_name] = pd.to_numeric(data[numeric_column_name], errors='coerce')

# 2. Type Conversion
# Convert the column to a numeric data type (e.g., float64)
data[numeric_column_name] = data[numeric_column_name].astype(float)

# Check for missing values after cleaning and conversion
missing_values_in_numeric_column = data[numeric_column_name].isnull().sum()
print(f"Missing values in '{numeric_column_name}': {missing_values_in_numeric_column}")
````

````python
mean_value = data[numeric_column_name].mean()
median_value = data[numeric_column_name].median()
std_deviation = data[numeric_column_name].std()
````

````python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data[numeric_column_name], bins=20)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Numeric Column')
plt.show()

# Create a box plot
plt.boxplot(data[numeric_column_name])
plt.ylabel('Values')
plt.title('Box Plot of Numeric Column')
plt.show()
````

````python
import pandas as pd
import numpy as np

# Load the original CSV
data = pd.read_csv('Alandi.csv', skiprows=4, header=None)

# Manually set meaningful column names
data.columns = ["Parameter", "Unnamed1", "Unnamed2", "Unnamed3", "Unnamed4", "Unnamed5", "Unnamed6"]

# Identify columns with missing values (NaN)
columns_with_missing_values = data.columns[data.isna().any()].tolist()

# Choose a statistical method (mean) for imputation
imputation_method = 'mean'

# Apply the selected statistical method to fill missing values
if imputation_method == 'mean':
    data[columns_with_missing_values] = data[columns_with_missing_values].apply(pd.to_numeric, errors='coerce')
    data[columns_with_missing_values] = data[columns_with_missing_values].fillna(data[columns_with_missing_values].mean())

# Remove rows with remaining NaN values
data_cleaned = data.dropna()

# Save the updated DataFrame to a new CSV file
data_cleaned.to_csv('Alandi_cleaned.csv', index=False)

# Display the first few rows of the cleaned data (optional)
print("\nCleaned Dataset:")
print(data_cleaned.head())

````
