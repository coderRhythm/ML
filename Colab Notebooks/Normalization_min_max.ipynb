```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load the CSV data into a pandas DataFrame
df = pd.read_csv("Preprocessed World University Rankings 2023 Dataset.csv")

# Select the columns you want to normalize (numeric columns)
numeric_columns = df.select_dtypes(include=['number'])

# Perform Min-Max normalization
min_max_scaler = MinMaxScaler()
df[numeric_columns.columns] = min_max_scaler.fit_transform(numeric_columns)

# Perform Z-score normalization
z_score_scaler = StandardScaler()
df[numeric_columns.columns] = z_score_scaler.fit_transform(numeric_columns)

# Save the normalized DataFrame to a new CSV file
df.to_csv("normalized_data.csv", index=False)
```
