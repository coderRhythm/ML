***Set***

```python
ThisSet = {1,1,2,3}
print(ThisSet)
```

```python
# Using add() function
ThisSet.add(4)
print(ThisSet)

#Using update() to add another set in a set
AnotherSet = {5,6,7}
AnotherSet.update(ThisSet)
print(AnotherSet)
```

```python
# remove the element in a set in python
ThisSet = {1,2,3,4}
ThisSet.remove(2)
print(ThisSet)
# remove() function will give an error if the item not present in set


#discard() function
ThisSet.discard(3)
print(ThisSet)

# Using pop() function
x = ThisSet.pop()
print("{} element will be poped".format(x))
print(ThisSet)


# Using clear function
print(ThisSet.clear())

# Using del keyword
thisset = {"apple", "banana", "cherry"}
del thisset
print(thisset)
```
