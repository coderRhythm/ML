```python
import tensorflow as tf
```

There are different Types of tensors:
zero dim , one dim, two dim, three dim

```python
# Zerp dimensional tensor
tensor_zero_d = tf.constant(4)
print(tensor_zero_d)
```

```python
# One dimensional tensor
tensor_one_d = tf.constant([1,2,3])
print(tensor_one_d)
```

```python
# Two dimensional tensor
tensor_two_d = tf.constant([[1,2,3],[4,5,6]])
print(tensor_two_d)
```

```python
# Three dimensional tensor
tensor_three_d = tf.constant([[[1,2,3],[4,5,6]],[[11,22,33],[44,55,66]]])
print(tensor_three_d)
```

```python
# To obtain tensor shape
# we use .shape()
print(tensor_three_d.shape)
```

```python
# finding the what is the type of tensor whether it is 1d,2d,3d
# we use .ndim
print(tensor_three_d.ndim)
```

```python
# four dimensional tensor
tensor_4_d = tf.constant([[[[1,2,3],[4,5,6]]],[[[11,22,33],[44,55,66]]],[[[1,2,3],[4,5,6]]],[[[11,22,33],[44,55,66]]]])
print(tensor_4_d)
```

```python
# eye method
# In eye method simple brackets are used for opening and closing
eye_tensor = tf.eye(
    3,  #when we assign no of rows = 3 then
    num_columns=None, # no of columns automatically 3
    batch_shape=None,
    dtype=tf.dtypes.float32,
    name=None
)

print(eye_tensor)

print('identical martrix * 3 : ')
print('')
print(3*eye_tensor)
```

```python
Another_eye_tensor = tf.eye(
    num_rows=5,
    num_columns=None,
    batch_shape=[2,],
    dtype = tf.dtypes.int16,
    name = None
)

print(Another_eye_tensor)
```

```python
# if we use number of columns which is not  same as num of rows
# then 1 is appear until the no of columns = no of rows
eye_tensor_not = tf.eye(
    num_rows = 5,
    num_columns = 3,
    batch_shape=None,
    dtype = tf.dtypes.int32,
    name = None
)

print(eye_tensor_not)
```

```python
# filling with scaler value
scalerTensor = tf.fill(
    [3,3],5,
    #first we specify dimensions then a value to be filled with dimesions
    #in this [2,3] first we specify no of rows and 3 which means 3 columns
    # the value to be fill is 5
    name = None
)

print(scalerTensor)


# In the advancement of this we have ones method
one_tensor = tf.ones(
    [2,3],
    dtype = tf.dtypes.int32,
    name = None
)
print('\nthis is one tensor')
print(one_tensor)

# if we make all elements to be zeroes then:
zero_tensor = tf.zeros(
    [3,3]
)

print(zero_tensor)
```

```python
# one_likeTensor = tf.ones_like(
#     scalerTensor
#     # name=None,
#     # dtype = tf.dtypes.int32
# )
one_likeTensor = tf.ones_like(scalerTensor)

print(one_likeTensor)

```

```python
#  finding the rank
tf.rank(scalerTensor)
#  print(Rank_tensor)
# we can find the rank of any dimension of tensor
# output as numpy  = 2

#calculate the size of the tensor matrix i.e the number of elements
print("The size is:")
tf.size(scalerTensor)
```

```python
# Creating the random tensor
Random_tensor = tf.random.normal(
    [3,3],
    mean = 100,
    stddev=1.0,
    dtype=tf.dtypes.float32,
    seed=None,
    name=None
)

print(Random_tensor)

print("This is the uniform tensor")
Uniform_tensor = tf.random.uniform(
    [3,3]
    # minval =0,
    # maxval = 8,
    # dtype = tf.dtypes.int32,  #if we use int32 then we must specify max value
    # seed = None,
    # name=None
)
print(Uniform_tensor)

# we use seed value so that when we call same random matrix with the same value
```

```python
# mathmatical operations
# absolute method
tensor_ = tf.constant([[-1,2,3],[-4,-6,5]])
print('before absolute : the tensor :')
print(tensor_)
print('')
print('after the absolute function: ')
abs_tensor = tf.abs(tensor_)
print(abs_tensor)
```

```python
sqrt_tensor = tf.sqrt((2.25)**2 + (4.75)**2)
print(sqrt_tensor)
```

```python
# adding the tensor
tensor_1 = tf.constant([1,2,3])
tensor_2 = tf.constant([4,5,6])

add_tensor = tf.add(tensor_1,tensor_2)
print(add_tensor)
```

```python
print(tf.multiply(tensor_1,tensor_2))
print(tf.divide(tensor_1,tensor_2))
```

```python
# using math library
# after adding .math into the existing statement
# the smaller tensor will stretched out to become the shape of bigger tensor
# i.e tf.constant([7]) is added to the tensor_1(which is the bigger tensor)
# for adding, we know that shape should be same
# so, tf.constant([7]) is going to streched out with the same element in repeated way to become the shape of tensor_1
# so In backend , after stretching  the tf.constant([7])--> [7,7,7]
# then it will be add
# we can print shape of the second tensor
print('the shape of first tensor then add')
print(tf.constant([7]).shape)
print("the tensor_1 is :")
print(tensor_1)
print('')
print('after adding:')
print(tf.math.add(tf.constant([7]),tensor_1))
```

```python
#now lets take another example
x_1 = tf.constant([[5,3,6,6,4,6]])
x_2 = tf.constant([[7],[5],[3]])
# it is element by element multiplication, not a matrix
print(tf.math.multiply(x_1,x_2))
```

```python
#  returns the max of two tensors
print('the maximum of two tensors:')
print(tf.math.maximum(tensor_1,tensor_2))
print('the minimum of two tensors')
print(tf.math.minimum(tensor_1,tensor_2))
```

```python
# we find the index of the maximum value using argmax
print(tf.math.argmax(tf.constant([100,2,3])))
```

```python
# we can find the index of maximum value of each column of 2d tensor using argmax
# In column , we set the axis  = 0
print(tf.math.argmax(tf.constant([[1,2,3],[4,5,6],[7,8,9]])),0)
# if we find the index with respect to row, we set the axis =1
print(tf.math.argmax(tf.constant([[1,2,3],[4,5,6],[7,8,9]])),1)
# we use reduce_min or reduce_max for finding the minimum and max of whole tensor, explore it!

```
