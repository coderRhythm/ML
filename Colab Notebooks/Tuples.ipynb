***Tuples***



```python
FirstTuple = (1,2,3)
#we can't change the value of the tuples
# FirstTuple[1] = 4  #This gives the error
print(FirstTuple)

SecondTuple = ('rhythm','age',19)
print(SecondTuple)


#Finding Number of elements of the tuple
print(len(SecondTuple))

#finding the type
print(type(SecondTuple))

ThirdTuple = (True,False,True)
print(ThirdTuple)


#Access Tuples
print('')
print('ThirdTuple elements\n')
for x in ThirdTuple:
  print(x)
```

**changing anf**

```python
#To change the value of the tuple , we have to convert into list
Tuple_list = (1,2,3,4)
new_list = list(Tuple_list)
new_list[1] = 5
print('initial Tuple is ')
print(Tuple_list)
Tuple_list = tuple(new_list)
print('after changing value of 2nd element ')
print(Tuple_list)


#we can't append the element in the tuple
#so we first convert tuple into list then append the list and then convert into tuple
New_Tuple_list = (1,2,3,4)
print('\ninitial tuple is ')
print(New_Tuple_list)
New_list = list(New_Tuple_list)
New_list.append(5);
New_Tuple_list = tuple(New_list)
print("after appending ")
print(New_Tuple_list)



#we can add tuple to tuple
print('\nadding tuple to tuple ')
FirstTuple = ("rhythm",)
SecondTuple = ("sethiya",)

FirstTuple+=SecondTuple
print(FirstTuple)
```

***Packing and Unpacking of a tuple ***

```python
One_tuple = ('rhythm','is','good','boy') #This is called Packing of  a tuple
print('\nthis is by packing')
print(One_tuple)
(name,verb,gooding,student) = One_tuple  #This is called Unpacking of a tuple
print("\nthis is by unpacking")
print(name,verb,gooding,student)
#In unpacking of a tuple , again Tuple's values's are assigned to the some other variable's

# we can also pass the list in a variable using star(asterisk(*))
(name,verb,*others) = One_tuple

print(name,verb,others)

```

```python
#Looping in Tuples
print('This:')
for x in One_tuple:
  print(x,end=' ')

#Through indexing
print('\nThis is through indexing')
thistuple = ('apple','banana','cherry')
for i in range(len(thistuple)):
  print(thistuple[i])


#Using while loop
i=0
while i<len(thistuple):
  print(thistuple[i])
  i=i+1

#Using the
```
