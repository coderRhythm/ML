```python
import pandas as pd

df = pd.read_csv('store_data.csv')
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming df is your original DataFrame
print(df.head())

# Fill NaN values with an empty string
df = df.fillna('')

# Create a list of transactions
transactions = df.values.tolist()

# Perform one-hot encoding using TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Apriori Algorithm
min_support = 0.05  # Adjusted min_support
current_support = 1.0  # Initialize with a value greater than min_support

while current_support >= min_support:
    # Find frequent itemsets
    frequent_itemsets = apriori(df_encoded, min_support=current_support, use_colnames=True)

    # Display frequent itemsets
    print(frequent_itemsets)

    # Generate association rules
    rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)

    # Display the generated rules
    print(rules)

    # Update current support for the next iteration
    if not frequent_itemsets.empty:
        current_support = frequent_itemsets['support'].min() - 0.01  # decrease by a small value
    else:
        break  # exit the loop if no frequent itemsets are found
```
