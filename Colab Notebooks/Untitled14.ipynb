```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
y = np.array([2, 4, 5, 4, 5])  # Dependent variable

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
new_X = np.array([6, 7, 8]).reshape(-1, 1)
predictions = model.predict(new_X)

# Print the coefficients and intercept
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Print the predictions
print("Predictions:", predictions)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Generate a range of X values for plotting
X_plot = np.linspace(min(X), max(X), 100).reshape(-1, 1)

# Make predictions for the entire range of X values
y_plot = model.predict(X_plot)

# Make predictions for new X values
new_X = np.array([6, 7, 8]).reshape(-1, 1)
predictions = model.predict(new_X)

# Print the coefficients and intercept
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Plot the data points
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot the linear regression line
plt.plot(X_plot, y_plot, color='red', linewidth=2, label='Linear Regression Line')

# Plot the predictions on new X values
plt.scatter(new_X, predictions, color='green', marker='o', label='Predictions')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.show()
```
