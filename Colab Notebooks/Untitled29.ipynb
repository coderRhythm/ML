```python
data = pd.read_csv("store_data.csv",encoding= "latin-1")
```

```python
data
```

```python
data.fillna('')
```

```python
te = TransactionEncoder()
```

```python
i = te.fit(data).transform(data)
```

```python
new_i = pd.DataFrame(i,columns = te.columns_)
```

```python
new_i
```

```python
from mlxtend.frequent_patterns import apriori
frequent_items = apriori(new_i,min_support = 0.001,use_colnames=True)
```

```python
frequent_items
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_items,metric="confidence",min_threshold = 0.007)
```
