```python
dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]
```

```python
data = np.array(dataset)
```

```python
data
```

```python
mean = np.mean(data)
```

```python
mean
```

```python
std = np.std(data)
```

```python
std
threshold  = 3
outliers = []
```

```python
for x in data:
  z_score = (x-mean)/std
  if z_score>threshold:
    outliers.append(x)
```

```python
outliers
```

```python
outliers
```

```python
min = np.percentile(data,25)
max = np.percentile(data,75)
threshold = 1.5
iqr = max-min
```

```python
lower_bound  = min-(threshold*iqr)
```

```python
upper_bound = max+(threshold*iqr)

outliers = []
```

```python
for x in data:
  if x<lower_bound or x>upper_bound :
    outliers.append(x)
```

```python
outliers
```
