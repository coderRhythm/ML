```python
import pandas as pd

data = {
  "calories": [420, 380,420, 390],
  "duration": [50, 40,50, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)

m = df['calories'].mean()
print(m)

median = df['duration'].median()
print(median)

m1  = df['duration'].mean()

var = ((df['calories']-m)**2).sum()/(len(df['calories'])-1)
print(var)


var1 = ((df['duration']-m1)**2).sum()/(len(df['duration'])-1)
print(var1)

#Range
r = df['calories'].max()-df['calories'].min()
print(r)

cov = ((df['calories']-m)*(df['duration']-m1)).sum()/(len(df['calories'])-1)
print(cov)

# Standard dev
std = df['calories'].std()
print(std)

std = df['duration'].std()
print(std)


#IQR
iqr  = df['calories'].quantile(0.75)-df['calories'].quantile(0.25)
print(iqr)


# correlation
corr = df.corr()
print(corr)
```

```python
# checking for duplication
v = df.duplicated(keep=False).values_count()
print(v)
print(df)
print(df.duplicated(keep=False))
```
