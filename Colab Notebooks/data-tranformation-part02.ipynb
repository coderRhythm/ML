```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define your dataset
dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]

# Calculate the mean using NumPy
mean = np.mean(dataset)

# Calculate the standard deviation
std_dev = np.std(dataset)

# Define a threshold for identifying outliers (commonly 2 or 3)
threshold = 3

# Identify and print outliers using a for loop
outliers = []
for x in dataset:
    z_score = abs((x - mean) / std_dev)
    if z_score > threshold:
        outliers.append(x)

print("Identified outliers:", outliers)
```

```python
import numpy as np

# Sample data (replace this with your data)
data = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]

# Step 2: Calculate the IQR (Interquartile Range)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Step 4: Define a threshold for identifying outliers (threshold = 3)
threshold = 3.0

# Step 5: Calculate the lower and upper bounds
lower_bound = Q1 - (threshold * IQR)
upper_bound = Q3 + (threshold * IQR)

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Step 6: Identify and display the outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Outliers:", outliers)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a DataFrame named df

# Select only the numeric columns
numeric_columns = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = np.corrcoef(numeric_columns.to_numpy(), rowvar=False)

# Create a heatmap
plt.figure(figsize=(15, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap for Top 5 Rows')
plt.show()
```
