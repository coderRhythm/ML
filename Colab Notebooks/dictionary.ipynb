```python
MyDict  = {
    "name":"Rhythm",
    "age":19,
    "college":"Mit wpu"
}

print(MyDict)


print(len(MyDict))


#dictionary values have unique key
#none two values have same key
#duplicate value overwrite its previous value
ThisDict = {
    'name':'Rhythm sethiya',
    'age':19,
    'age':'nineteen',
    'college':"mit wpu",
    "goodcollge":False,
    'Subjects':('operating system','python','C and C++','Dec')
}
print(ThisDict)
```

We can make dictionary using dict keyword

```python
Mydict = dict(
              name="rhythm",
              college="Mit wpu",
              GoodCollege=False
            )
print(Mydict)
```

***Access items in dictionary ***

```python
#accessing values using keys
print('Acessing keys using index method')
print(Mydict['name'])
print()

#using loop
print('Through loop')
for i in Mydict:  # 'i' is basically keys in dictionary
  print(Mydict[i])

#we can access values using .get() function
print('\naccesing values using .get() function')
x = Mydict.get('name')
print(x)
print()

#using loop and .get() function
print('using loop and .get() function ')
for i in Mydict:
  x = Mydict.get(i)
  print(x)

# returning all the keys in the dctionary
print('\nLets see all the keys:')
print(Mydict.keys())

print('lets see all the values ')
print(Mydict.values())


#adding items to the dictionary
print('before:')
print(Mydict)
print('After')
Mydict['age'] = 19
print(Mydict)


#returning each items as a tuple in a dictionary
```

```python
# Looping in dictionary
print('Printing keys')
for i in Mydict.keys():
  print(i)

print('\nprinting values')
for i in Mydict.values():
  print(i)

print('printing keys and its corresponding values')
for i,j in Mydict.items():
  print(i,':',j)
```

***Methods in dictionary ***

```python
# Copy methods
Copy_dict = Mydict.copy();
print(Copy_dict)

copy_Dict = dict(Mydict)
print(copy_Dict)
```

**Nested Dictionary**

```python
#implimentation of nested dictionary
nested_dict  = dict(
    dict1 = dict(name='Rhythm sethiya',age=19,college='mit wpu'),
    dict2 = dict(name='harsh jain',age = 20,college='mit wpu'),
    dict3 = dict(name='kavish jain',age=20,college='mit wpu'),
    dict4 = dict(name='sarthak jagetiya',age=20,college='mit wpu')
)
print(nested_dict)

print(nested_dict['dict1'])
```
