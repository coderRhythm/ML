***SERIES***
1. A Pandas Series is indeed a one-dimensional array-like data structure that can hold data of any type.
2.  It can be thought of as similar to a column in a table, but it's not limited to being used only as a representation of a table column. You can create a Series from various data collections like lists, arrays, or dictionaries.

```python
import pandas as pd
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)

ThisTuple = ('rhythm','aryanm','pranav','harsh','akarsh')
sere = pd.Series(ThisTuple)
print(sere)

ThisDict = {'name1:':'rhythm','name2:':'pranav'}
serdict = pd.Series(ThisDict)
print(serdict)


ThisState = {'maharashtra':'mumbai','rajasthan':'jaipur','bihar':'patna','gujarat':'gandhinagar'}
seredict = pd.Series(ThisState)
print(seredict)
import numpy as np
arr = np.array([1,2,3])
print(arr)
```

` When creating a Series, we can also pass an optional index attribute to specify custom index labels for the values. This allows  to access the elements using these index labels, and it's particularly useful when we want to provide meaningful labels or identifiers for the data points.`

```python
data_list = [10, 20, 30, 40, 50]
series_from_list = pd.Series(data_list,index=[5,6,7,8,9])
print(series_from_list)



#index taking string
IndexString = ['n1','n2','n3','n4','n5']
print(series_from_list,IndexString)
```

***DataFrame***

A DataFrame is a two-dimensional, size-mutable, and heterogeneous tabular data structure in Pandas. It's designed to store and manipulate data in a way that's similar to a spreadsheet or a SQL table. *We* can think of a DataFrame as a collection of Pandas Series, where each Series represents a column and is aligned by index labels to form the rows of the DataFrame.

```python

DataDict = {
    1: {
        'name': 'rhythm',
        'age': 19,
        'college': 'mit wpu'
    },
    2: {
        'name': 'Aryan sharma',
        'age': 20,
        'college': 'mit wpu'
    },
    3: {
        'name': 'akarsh singh',
        'age': 21,
        'college': 'mit wpu'
    }
}

df = pd.DataFrame(DataDict)

print(df)



```

Accessing elements in a dataframe using two methods:
M-1: Using index:  
  

```python
for key in DataDict.keys():
    pdf = pd.DataFrame(DataDict)
    print(pdf[key]['name'])
```

***Arithematic operations***

```python
#Arithematic operations
ser1 = pd.Series([1,2,3])
ser2 = pd.Series([4,5,6])

print(ser1+ser2)
print(ser1*ser2)
print(ser1-ser2)
print(ser1/ser2)
print(ser1%ser2)

```
