# STATISTICAL ANALYSIS

```python
import pandas as pd

# Load the actual dataset
df = pd.read_csv('Charades_v1_clean_train_100.csv')

# Display DataFrame
print("Actual DataFrame:")
print(df)
print("\n")

# Summary statistics for numerical columns
numerical_summary = df[['quality', 'relevance', 'length']].describe()

# Distribution of scenes, subjects, and categories
scene_distribution = df['scene'].value_counts()
subject_distribution = df['subject'].value_counts()

# Verification status distribution
verification_distribution = df['verified'].value_counts()

# Most common objects, descriptions, and actions
common_objects = df['objects'].str.split(';').explode().value_counts()
common_descriptions = df['descriptions'].str.split(';').explode().value_counts()
common_actions = df['actions'].str.split(';').explode().value_counts()

# Display results
print("Summary Statistics for Numerical Columns:")
print(numerical_summary)
print("\nDistribution of Scenes:")
print(scene_distribution)
print("\nDistribution of Subjects:")
print(subject_distribution)
print("\nVerification Status Distribution:")
print(verification_distribution)
print("\nMost Common Objects:")
print(common_objects.head(10))
print("\nMost Common Descriptions:")
print(common_descriptions.head(10))
print("\nMost Common Actions:")
print(common_actions.head(10))
```

# **Object wise dataframe**

```python
sub_rows = []
for index, row in df.iterrows():
    objects = row['objects'].split(';')
    actions = row['actions'].split(';')
    for obj in objects:
        for action in actions:
            sub_row = row.copy()
            sub_row['objects'] = obj
            sub_row['actions'] = action
            sub_rows.append(sub_row)
new_df = pd.DataFrame(sub_rows)

```
